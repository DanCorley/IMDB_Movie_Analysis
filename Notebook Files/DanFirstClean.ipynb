```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
bom = pd.read_csv('bom.movie_gross.csv')
tn = pd.read_csv('tn.movie_budgets.csv')
imdb = pd.read_csv('imdb.title.basics.csv')
imdb_rate = pd.read_csv('imdb.title.ratings.csv')
```

```python
imdb.sample(5)
```

```python
tn['worldwide_gross'] = tn['worldwide_gross'].replace('[\$,]','',regex=True).astype('int64')
tn['domestic_gross'] = tn['domestic_gross'].replace('[\$,]','',regex=True).astype('int64')
tn['production_budget'] = tn['production_budget'].replace('[\$,]','',regex=True).astype('int64')
```

```python
tn.sort_values(by='domestic_gross', ascending=False, inplace=True)
```
